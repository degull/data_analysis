```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 데이터프레임에서 열 선택
#df_selected = df[['Axis [nm]', 'ROI 1 []', 'ROI 2 []', 'ROI 3 []', 'ROI 4 []', 'ROI 5 []', 'ROI 6 []', 'ROI 7 []', 'ROI 8 []', 'ROI 9 []', 'ROI 10 []', 'ROI 11 []', 'ROI 12 []', 'ROI 13 []']]

# 데이터프레임을 행렬로 변환
data_matrix = df_selected.values[:, 1:].astype(float).T

# 히트맵 그리기
fig, ax = plt.subplots(figsize=(30, 6))
im = ax.imshow(data_matrix, cmap='hot', aspect='auto')

# x축, y축 설정
ax.set_xticks(np.arange(len(df_selected)))
ax.set_yticks(np.arange(len(df_selected.columns)-1))
ax.set_xticklabels(df_selected['Axis [nm]'], rotation=45, ha='right')
ax.set_yticklabels(df_selected.columns[1:])

# 각 셀에 값 표시
for i in range(len(df_selected.columns)-1):
    for j in range(len(df_selected)-1):
        text = ax.text(j, i, f'{data_matrix[i, j]:.1f}', ha='center', va='center', color='w')

# 그래프 제목 설정
plt.title('ROIs의 히트맵')

# 컬러바 추가
cbar = ax.figure.colorbar(im)

# 그래프 표시
plt.show()
```
