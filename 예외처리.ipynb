```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 주어진 데이터를 DataFrame으로 변환
data = {
    'Axis [nm]': [862.259300, 862.459191, 862.659082, 862.858974, 863.058865, 863.258756, 863.458647, 863.658538, 863.858430, 864.058321, 864.258212, 864.458103, 864.657994, 864.857885, 865.057777, 865.257668, 865.457559, 865.657450, 865.857341, 866.057233],
    'ROI 1': [10.612855, 10.796809, 11.248751, 11.016404, 11.033107, 11.313623, 10.820744, 12.28849452, 10.781374, 10.949146, 11.267467, 12.974797, 13.822100, 13.822832, 13.191634, 14.252106, 14.389531, 14.464171, 15.265528, 15.197524],
    'ROI 2': [10.812855, 10.396809, 11.848751, 12.016404, 10.033107, 11.313623, 10.820744, 12.88849452, 9.781374, 11.949146, 11.267467, 12.174797, 13.522100, 13.122832, 13.891634, 14.852106, 14.189531, 14.264171, 15.565528, 15.397524]
}
df = pd.DataFrame(data)

# Baseline 잡기
def baseline_als(y, lam, p, niter=100):
    L = len(y)
    D = np.diff(np.eye(L), 2)
    w = np.ones(L)
    for i in range(niter):
        W = np.diag(w)
        Z = W + lam * D.dot(D.transpose())
        z = np.linalg.solve(Z, w*y.values)  # Convert y to numpy array using values
        w = p * (y > z) + (1-p) * (y < z)
    return z

# Baseline 계산
baseline = baseline_als(df['ROI 1'], lam=100, p=0.1)  # Use a higher lam value for a smoother baseline

# 그래프 출력
plt.plot(df['Axis [nm]'], df['ROI 1'] - baseline, label='ROI 1 - Baseline')  # Subtract baseline from ROI 1
plt.plot(df['Axis [nm]'], df['ROI 2'] - baseline, label='ROI 2 - Baseline')  # Subtract baseline from ROI 2
plt.plot(df['Axis [nm]'], np.zeros_like(df['Axis [nm]']), '--', color='black')  # Straight line at y=0 (baseline)
plt.xlabel('Axis [nm]')
plt.ylabel('Value')
plt.title('Line Graph')
plt.legend()
plt.grid(True)
plt.show()
```
