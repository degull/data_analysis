# 링크연결

```python
import sys
from PyQt5.QtWidgets import QVBoxLayout, QWidget

from PyQt5.QtWidgets import QApplication, QMainWindow, QTextBrowser, QPushButton, QFileDialog
from PyQt5.QtGui import QDesktopServices
from PyQt5.QtCore import QUrl, Qt
from openpyxl import load_workbook

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Excel Viewer")
        self.setGeometry(100, 100, 400, 300)

        self.text_browser = QTextBrowser(self)
        self.setCentralWidget(self.text_browser)

        self.load_button = QPushButton("불러오기", self)
        self.load_button.clicked.connect(self.load_button_clicked)

        self.data = ""  # 선택한 데이터를 저장할 변수

        layout = QVBoxLayout()
        layout.addWidget(self.text_browser)
        layout.addWidget(self.load_button)

        central_widget = QWidget(self)
        central_widget.setLayout(layout)
        self.setCentralWidget(central_widget)

    def load_button_clicked(self):
        file_dialog = QFileDialog(self)
        file_path, _ = file_dialog.getOpenFileName(self, "엑셀 파일 불러오기")
        if file_path:
            self.load_excel(file_path)

    def load_excel(self, file_path):
        workbook = load_workbook(file_path)
        sheet = workbook.active

        rows = sheet.max_row
        columns = sheet.max_column

        data = ""
        for row in range(1, rows + 1):
            for col in range(1, columns + 1):
                cell = sheet.cell(row=row, column=col)
                cell_value = cell.value
                if cell.hyperlink:
                    hyperlink = cell.hyperlink.target
                    cell_value = f'<a href="{hyperlink}">{str(cell_value)}</a>'
                data += str(cell_value) + "\t"
            data += "<br>"

        self.text_browser.setHtml(data)
        self.data = data  # 데이터를 변수에 저장

    def anchorClicked(self, url):
        self.link_clicked(url.toString())

    def link_clicked(self, url):
        # 링크 클릭 시 웹 브라우저로 연결
        QDesktopServices.openUrl(QUrl(url))

if __name__ == "__main__":
    app = QApplication(sys.argv)
    app.setDesktopSettingsAware(False)  # 데스크탑 설정 비활성화
    app.setAttribute(Qt.AA_EnableHighDpiScaling, True)  # 고해상도 스케일링 활성화
    window = MainWindow()
    window.show()
    sys.exit(app.exec_())

```
